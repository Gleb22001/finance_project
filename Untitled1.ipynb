```python
import numpy as np
import pandas as pd
data='Rates.xlsx'
data=pd.ExcelFile(data)
print(data.sheet_names)
df = data.parse('Data')
df['spread_1y']=df['s_1y']-df['r_1y']
df
```

# Предваритеьльный анализ

```python
from matplotlib import pyplot as plt   
#ploting our canvas   
plt.plot(df['Dates'], df['s_1y'], color='blue', label="swap")  
plt.plot(df['Dates'], df['r_1y'], color='red', label="odz")
plt.legend(loc='upper left', frameon=False)
#display the graph   
plt.show()  
```

# Анализ на основе копул

```python
import copulas
from copulas.multivariate import GaussianMultivariate
from copulas.visualization import compare_2d
data=df[['s_1y', 'r_1y']]
copula=GaussianMultivariate()
copula.fit(data)
synthetic_data=copula.sample(len(data))
compare_2d(data,synthetic_data)
```

```python
from copulas.visualization import compare_3d
data1=df[['s_1y', 's_2y', 's_3y']]
copula=GaussianMultivariate()
copula.fit(data1)
synthetic_data1=copula.sample(len(data1))
compare_3d(data1,synthetic_data1)
```
